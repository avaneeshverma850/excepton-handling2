```python
class ArithmeticError(Exception):
    pass
try:
    a=int(input("enter number:"))
    b=int(input("enter number:"))
    if b==0:
        raise ArithmeticError
    else:
        print(a/b)
except ArithmeticError:
    print("value of b cannot be zero")
```
